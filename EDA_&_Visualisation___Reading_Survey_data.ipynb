# Introduction

Surveys are one of the most common methods used to collect primary data, especially in understanding students’ experiences and perspectives. Through surveys, researchers are able to gather information directly from respondents regarding specific issues that may not be easily observed.

This project investigates the relationship between activities related to CV development and stress levels among university students. The survey focuses on students’ involvement in activities such as organizations, competitions, seminars, internships, and other experiences intended to support career preparation and improve their CVs.

However, survey-based research is also sensitive to several limitations, such as respondent bias, differences in individual interpretations of questions, and limited sample size. Therefore, this project also considers the reliability of the collected data and several factors that may influence the survey results.

# Overview
The dataset used in this study consists of responses collected from 35 university students through an online questionnaire. The survey focuses on students’ activities related to CV development, such as participation in organizations, competitions, seminars, internships, and other self-development activities, as well as the stress experienced by students during these activities.

The dataset contains respondent demographic information, including age, gender, study program, and semester level. In addition, the questionnaire includes several Likert-scale questions related to CV development activities and student stress levels.

Since the survey was conducted using close-ended questions and Likert scales, respondents selected answers based on predefined options. While this approach simplifies data collection and analysis, it also limits the depth of respondents’ explanations and experiences regarding the topic.

```python
x_columns = [
    "Saya merasa tertekan karena harus memiliki CV yang lengkap dan menarik sebelum lulus. ",
    "Saya khawatir pengalaman organisasi atau kepanitiaan saya tidak cukup untuk dimasukkan ke CV.",
    "Saya khawatir pengalaman organisasi atau kepanitiaan saya tidak cukup untuk dimasukkan ke CV.",
    "Saya merasa teman-teman memiliki pencapaian yang lebih banyak dibanding saya (misalnya magang, proyek, sertifikat).",
    "Saya merasa ada tuntutan eksternal untuk mengambil banyak kegiatan demi mempercantik CV.",
    "Saya sering membandingkan CV saya dengan CV orang lain (dari media sosial, LinkedIn, dll.).",
    "Tekanan untuk membangun CV terasa meningkat menjelang semester akhir atau masa rekrutmen kerja.",
    "Saya menghabiskan waktu yang sangat lama hanya untuk memikirkan atau memperbaiki tampilan CV."
]


raw_df["CV Score"] = raw_df[x_columns].mean(axis=1)
```

```python
y_columns = [
    "Saya sering merasa cemas memikirkan masa depan karier saya.",
    "Saya merasa cemas ketika tidak mengikuti kegiatan yang dapat menambah pengalaman saya.",
    "Saya merasa kewalahan dalam menjalani berbagai kegiatan seperti organisasi, lomba, atau pelatihan.",
    "Saya merasa kewalahan dengan banyaknya hal yang harus disiapkan selama kuliah.",
    "Saya merasa terbebani dengan banyaknya kegiatan yang saya jalani di luar perkuliahan.",
    "Saya merasa lelah secara fisik atau pusing ketika memikirkan target pencapaian yang belum tercapai.",
    "Saya merasa stres ketika melihat teman yang sudah mendapat pengalaman kerja atau magang.",
    "Kegiatan tambahan yang saya jalani mempengaruhi kondisi mental saya.",
    "Saya merasa takut tidak siap saat nanti terjun ke dunia kerja."
]

raw_df["Stress Score"] = raw_df[y_columns].mean(axis=1)
```

```python
correlation = raw_df["CV Score"].corr(raw_df["Stress Score"])

print("Correlation:", correlation)
```

A correlation value of 0.68 indicates a strong positive relationship between CV development activities and student stress levels. This suggests that the greater the student's involvement in CV development activities, the higher their stress levels tend to be.

```python
# import library
import plotly.express as px

# create interactive scatter plot
fig = px.scatter(
    raw_df,
    x="CV Score",
    y="Stress Score",
    hover_data=["Semester", "Jenis Kelamin"],
    title="Relationship between CV Development Activities and Student Stress Levels",
    labels={
        "CV Score": "CV Development Score",
        "Stress Score": "Stress Level Score"
    }
)

# update layout
fig.update_layout(
    width=1000,
    height=600,
    title_font_size=20
)

# show graph
fig.show()
```

```python
import plotly.express as px
import pandas as pd

# kategorisasi score
raw_df["CV Category"] = pd.cut(
    raw_df["CV Score"],
    bins=[1, 2, 3, 4, 5],
    labels=["Very Low", "Low", "High", "Very High"]
)

raw_df["Stress Category"] = pd.cut(
    raw_df["Stress Score"],
    bins=[1, 2, 3, 4, 5],
    labels=["Very Low", "Low", "High", "Very High"]
)

# membuat crosstab
cv_stress_ct = pd.crosstab(
    raw_df["CV Category"],
    raw_df["Stress Category"]
)

# ubah ke format dataframe
heatmap_data = cv_stress_ct.reset_index().melt(
    id_vars='CV Category',
    var_name='Stress Category',
    value_name='Count'
)

# membuat interactive heatmap
fig = px.density_heatmap(
    heatmap_data,
    x="Stress Category",
    y="CV Category",
    z="Count",
    text_auto=True,
    color_continuous_scale="Purples",
    title="Relationship Between CV Development Activities and Stress Levels"
)

# update layout
fig.update_layout(
    width=900,
    height=600
)

# tampilkan grafik
fig.show()
```
