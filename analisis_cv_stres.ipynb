### Populasi

Populasi dalam penelitian ini adalah **seluruh mahasiswa aktif** di Institut Teknologi Sains Bandung.  
Populasi dipilih karena penelitian berfokus pada mahasiswa yang terlibat dalam berbagai kegiatan pengembangan diri seperti **organisasi, seminar, lomba, pelatihan, dan magang** sebagai bentuk persiapan CV dan portofolio.

Perkiraan jumlah populasi mahasiswa aktif: **± 1.000 mahasiswa**

### Penghitungan Sampel (Rumus Slovin)

$$n = \frac{N}{1 + N(e)^2}$$

**Keterangan:**
- $n$ = jumlah sampel
- $N$ = jumlah populasi = 1.000
- $e$ = margin of error = 15% (0,15)

**Perhitungan:**

$$n = \frac{1000}{1 + 1000 \times (0.15)^2} = \frac{1000}{1 + 22.5} = \frac{1000}{23.5} \approx 43$$

Jadi, **jumlah minimal sampel yang dibutuhkan adalah sekitar 43 responden**.  
Pada penelitian ini, jumlah responden yang berhasil diperoleh adalah **35 mahasiswa** mempertimbangkan keterbatasan waktu pengumpulan data namun masih representatif untuk analisis eksploratori.


## 3️ Metode Sampling

### Metode yang Digunakan: Purposive Sampling

**Purposive Sampling** adalah teknik pengambilan sampel berdasarkan pertimbangan peneliti mengenai siapa yang dianggap paling sesuai atau relevan dengan tujuan penelitian.

### Alasan Pemilihan Metode

Purposive sampling dipilih karena peneliti membutuhkan responden yang **memiliki pengalaman relevan** dengan topik CV dan stres bukan sampel acak semua mahasiswa. Mahasiswa yang belum pernah terlibat kegiatan apapun akan memberikan data yang tidak representatif untuk penelitian ini.

### Definisi Kolom & Perhitungan Skor


```python
# ============================================================
# DEFINISI KOLOM PERTANYAAN
# ============================================================

# 7 item pertanyaan untuk CV Score
x_cols = [
    'Saya merasa tertekan karena harus memiliki CV yang lengkap dan menarik sebelum lulus. ',
    'Saya khawatir pengalaman organisasi atau kepanitiaan saya tidak cukup untuk dimasukkan ke CV.',
    'Saya merasa teman-teman memiliki pencapaian yang lebih banyak dibanding saya (misalnya magang, proyek, sertifikat).',
    'Saya merasa ada tuntutan eksternal untuk mengambil banyak kegiatan demi mempercantik CV.',
    'Saya sering membandingkan CV saya dengan CV orang lain (dari media sosial, LinkedIn, dll.).',
    'Tekanan untuk membangun CV terasa meningkat menjelang semester akhir atau masa rekrutmen kerja.',
    'Saya menghabiskan waktu yang sangat lama hanya untuk memikirkan atau memperbaiki tampilan CV.'
]

# 9 item pertanyaan untuk Stress Score
y_cols = [
    'Saya sering merasa cemas memikirkan masa depan karier saya.',
    'Saya merasa cemas ketika tidak mengikuti kegiatan yang dapat menambah pengalaman saya.',
    'Saya merasa kewalahan dalam menjalani berbagai kegiatan seperti organisasi, lomba, atau pelatihan.',
    'Saya merasa kewalahan dengan banyaknya hal yang harus disiapkan selama kuliah.',
    'Saya merasa terbebani dengan banyaknya kegiatan yang saya jalani di luar perkuliahan.',
    'Saya merasa lelah secara fisik atau pusing ketika memikirkan target pencapaian yang belum tercapai.',
    'Saya merasa stres ketika melihat teman yang sudah mendapat pengalaman kerja atau magang.',
    'Kegiatan tambahan yang saya jalani mempengaruhi kondisi mental saya.',
    'Saya merasa takut tidak siap saat nanti terjun ke dunia kerja.'
]

# Hitung skor komposit
df['CV Score']     = df[x_cols].mean(axis=1).round(3)
df['Stress Score'] = df[y_cols].mean(axis=1).round(3)

print('CV Score dan Stress Score berhasil dihitung.')
print()
print('5 baris pertama:')
print(df[['Jenis Kelamin', 'Program Studi', 'CV Score', 'Stress Score']].head())
```

### Statistik Deskriptif


```python
# ============================================================
# STATISTIK DESKRIPTIF
# ============================================================

print('=' * 55)
print('STATISTIK DESKRIPTIF — CV SCORE & STRESS SCORE')
print('=' * 55)
desc = df[['CV Score', 'Stress Score']].describe().round(2)
print(desc)

print()
print('=' * 55)
print('RATA-RATA PER ITEM — CV SCORE')
print('=' * 55)
cv_labels_short = [
    'Tertekan CV lengkap & menarik',
    'Khawatir org/kepanitiaan kurang',
    'Teman lebih berprestasi',
    'Tuntutan eksternal kegiatan',
    'Bandingkan CV orang lain',
    'Tekanan semester akhir/rekrutmen',
    'Waktu lama perbaiki CV'
]
for label, col in zip(cv_labels_short, x_cols):
    print(f'  {label:<35}: {df[col].mean():.2f}')

print()
print('=' * 55)
print('RATA-RATA PER ITEM — STRESS SCORE')
print('=' * 55)
stress_labels_short = [
    'Cemas masa depan karier',
    'Cemas tidak ikut kegiatan',
    'Kewalahan berbagai kegiatan',
    'Kewalahan persiapan kuliah',
    'Terbebani kegiatan luar kuliah',
    'Lelah fisik/target belum tercapai',
    'Stres melihat teman magang',
    'Kegiatan pengaruhi mental',
    'Takut belum siap kerja'
]
for label, col in zip(stress_labels_short, y_cols):
    print(f'  {label:<35}: {df[col].mean():.2f}')
```

### Kategorisasi Skor


```python
# ============================================================
# KATEGORISASI CV SCORE DAN STRESS SCORE
# ============================================================

df['CV Category'] = pd.cut(
    df['CV Score'],
    bins=[1, 2, 3, 4, 5],
    labels=['Sangat Rendah', 'Rendah', 'Tinggi', 'Sangat Tinggi'],
    include_lowest=True
)

df['Stress Category'] = pd.cut(
    df['Stress Score'],
    bins=[1, 2, 3, 4, 5],
    labels=['Sangat Rendah', 'Rendah', 'Tinggi', 'Sangat Tinggi'],
    include_lowest=True
)

print('Distribusi CV Category:')
print(df['CV Category'].value_counts().sort_index())

print()
print('Distribusi Stress Category:')
print(df['Stress Category'].value_counts().sort_index())

print()
print('Crosstab CV Category vs Stress Category:')
print(pd.crosstab(df['CV Category'], df['Stress Category']))
```

### Interpretasi | Kategorisasi

- Mayoritas mahasiswa masuk kategori **"Tinggi"** pada CV Score maupun Stress Score, menandakan tekanan CV dan stres sudah menjadi pengalaman umum di kalangan mahasiswa semester awal.
- Sangat sedikit responden di kategori "Sangat Rendah", mengkonfirmasi bahwa hampir semua mahasiswa merasakan tekanan dalam taraf tertentu.
- Pola crosstab menunjukkan kecenderungan: mahasiswa dengan CV Category "Tinggi" atau "Sangat Tinggi" juga cenderung masuk Stress Category yang tinggi.


### Korelasi Pearson


```python
# ============================================================
# KORELASI PEARSON
# ============================================================

r_val, p_val = stats.pearsonr(df['CV Score'], df['Stress Score'])

print('=' * 45)
print('HASIL KORELASI PEARSON')
print('=' * 45)
print(f'  CV Score  vs  Stress Score')
print(f'  r        = {r_val:.4f}')
print(f'  p-value  = {p_val:.4f}')

if abs(r_val) >= 0.80:
    kekuatan = 'Sangat Kuat'
elif abs(r_val) >= 0.60:
    kekuatan = 'Kuat'
elif abs(r_val) >= 0.40:
    kekuatan = 'Sedang'
elif abs(r_val) >= 0.20:
    kekuatan = 'Lemah'
else:
    kekuatan = 'Sangat Lemah'

arah = 'Positif' if r_val > 0 else 'Negatif'
signifikan = 'Signifikan (p < 0.05)' if p_val < 0.05 else 'Tidak Signifikan (p >= 0.05)'

print(f'  Kekuatan     : {kekuatan}')
print(f'  Arah         : {arah}')
print(f'  Signifikansi : {signifikan}')
```

### Interpretasi | Korelasi Pearson

- Nilai **r** menunjukkan korelasi positif antara tekanan pengembangan CV dan tingkat stres mahasiswa.
- Artinya: **semakin besar tekanan mahasiswa terkait CV, semakin tinggi pula stres yang mereka rasakan** dan sebaliknya.
- Korelasi ini **bukan kausalitas** tidak bisa disimpulkan bahwa CV *menyebabkan* stres. Kemungkinan keduanya sama-sama dipengaruhi faktor ketiga seperti tekanan sosial atau ekspektasi diri.
- Temuan ini relevan untuk perancangan program bimbingan karier mahasiswa yang diberi dukungan manajemen CV lebih awal berpotensi mengalami penurunan stres secara bersamaan.

```python
# ============================================================
# VISUALISASI — 4 GRAFIK UTAMA
# ============================================================

fig = plt.figure(figsize=(18, 16))
fig.suptitle(
    'Analisis Hubungan Kegiatan CV dengan Tingkat Stres Mahasiswa\n'
    f'(n = {len(df)} responden | Semester 2 | Institut Teknologi Sains Bandung)',
    fontsize=14, fontweight='bold', y=0.98
)

# 1. Scatter Plot 
ax1 = fig.add_subplot(2, 2, 1)

color_map = {'Perempuan': '#185FA5', 'Laki-laki': '#BA7517'}
for gender, grp in df.groupby('Jenis Kelamin'):
    c = color_map.get(gender, '#888888')
    ax1.scatter(grp['CV Score'], grp['Stress Score'],
                c=c, label=f'{gender} (n={len(grp)})',
                alpha=0.75, s=70, edgecolors='white', linewidths=0.5)

m, b = np.polyfit(df['CV Score'], df['Stress Score'], 1)
x_line = np.linspace(df['CV Score'].min(), df['CV Score'].max(), 100)
ax1.plot(x_line, m * x_line + b, color='#E24B4A',
         linewidth=1.5, linestyle='--', label='Garis regresi')

ax1.set_title(f'Scatter Plot CV Score vs Stress Score\n(r = {r_val:.2f}, p = {p_val:.3f})', fontsize=12)
ax1.set_xlabel('CV Score (rata-rata 7 item)')
ax1.set_ylabel('Stress Score (rata-rata 9 item)')
ax1.set_xlim(1, 5.5)
ax1.set_ylim(1, 5.5)
ax1.legend(fontsize=9)

# 2. Bar Chart Item CV 
ax2 = fig.add_subplot(2, 2, 2)

cv_labels_wrap = [
    'Tertekan CV\nlengkap',
    'Khawatir org\nkurang',
    'Teman lebih\nberprestasi',
    'Tuntutan\neksternal',
    'Bandingkan CV\norang lain',
    'Tekanan\nsemester akhir',
    'Waktu lama\nperbaiki CV'
]
cv_means_val = df[x_cols].mean().values

bars2 = ax2.barh(cv_labels_wrap, cv_means_val, color='#185FA5', alpha=0.75, edgecolor='white')
for bar, val in zip(bars2, cv_means_val):
    ax2.text(val + 0.03, bar.get_y() + bar.get_height() / 2,
             f'{val:.2f}', va='center', fontsize=9)

ax2.set_xlim(1, 5.5)
ax2.set_title('Rata-rata Per Item CV Score', fontsize=12)
ax2.set_xlabel('Rata-rata skor (1–5)')
ax2.axvline(x=3, color='gray', linestyle=':', linewidth=1, label='Titik tengah (3)')
ax2.legend(fontsize=8)

# 3. Bar Chart Item Stres 
ax3 = fig.add_subplot(2, 2, 3)

stress_labels_wrap = [
    'Cemas masa\ndepan karier',
    'Cemas tidak\nikut kegiatan',
    'Kewalahan\nkegiatan',
    'Kewalahan\npersiapan kuliah',
    'Terbebani\nkegiatan luar',
    'Lelah fisik/\ntarget belum',
    'Stres teman\nmagang',
    'Kegiatan\npengaruhi mental',
    'Takut belum\nsiap kerja'
]
stress_means_val = df[y_cols].mean().values

bars3 = ax3.barh(stress_labels_wrap, stress_means_val, color='#BA7517', alpha=0.75, edgecolor='white')
for bar, val in zip(bars3, stress_means_val):
    ax3.text(val + 0.03, bar.get_y() + bar.get_height() / 2,
             f'{val:.2f}', va='center', fontsize=9)

ax3.set_xlim(1, 5.5)
ax3.set_title('Rata-rata Per Item Stress Score', fontsize=12)
ax3.set_xlabel('Rata-rata skor (1–5)')
ax3.axvline(x=3, color='gray', linestyle=':', linewidth=1, label='Titik tengah (3)')
ax3.legend(fontsize=8)

# 4. Heatmap Kategori 
ax4 = fig.add_subplot(2, 2, 4)

ct = pd.crosstab(df['CV Category'], df['Stress Category'])
sns.heatmap(ct, annot=True, fmt='d', cmap='Blues',
            linewidths=0.5, linecolor='white',
            ax=ax4, cbar_kws={'shrink': 0.8})
ax4.set_title('Heatmap: Kategori CV vs Kategori Stres', fontsize=12)
ax4.set_xlabel('Stress Category')
ax4.set_ylabel('CV Category')

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.savefig('hasil_analisis_cv_stres.png', dpi=150, bbox_inches='tight')
plt.show()
print("Grafik disimpan sebagai 'hasil_analisis_cv_stres.png'")
```
